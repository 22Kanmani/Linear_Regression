```python
# defining seed for consistent results
seed = 1000
np.random.seed(seed)
```

```python
# This is a pre-Defined evaluation function for calculating some helpful metrics. You may use it to evaluate your model.
# This function calculates the mean square error and the square root of the MSE for our prediction model.
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from math import sqrt

def evaluate (model, test_features, test_labels):
    predictions=model.predict(test_features)
    r2=r2_score(y_pred=predictions, y_true=test_labels)
    MAE=mean_absolute_error(test_labels, predictions)
    MSE=mean_squared_error(test_labels, predictions)
    rmse = sqrt(MSE)

    print('r2 = {:0.2f}'.format(r2))
    print('MAE = {:0.2f}'.format(MAE))
    print('Mean Square Error = {:0.2f}'.format(MSE))
    print('Squareroot Mean Square Error = {:0.2f}'.format(rmse))
    
    return round(r2,2)
```

```python
# Splitting data into Train and Test samples, 80% train data
from sklearn.model_selection import train_test_split

inp_features = updated_df['RegionID']
out_features = updated_df['2020-02']

x_train, x_test,y_train, y_test=train_test_split(inp_features,out_features,train_size=0.8,random_state= seed)

print('Shape of x_train = ',str(x_train.shape))
print('Shape of x_test  = ',str(x_test.shape))
print('Shape of y_train = ',str(y_train.shape))
print('Shape of y_test  = ',str(y_test.shape))
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
```

```python
linear_reg_model = LinearRegression()  
```

```python
x_train
```

```python
x_train= x_train.values.reshape(-1, 1)
x_test = x_test.values.reshape(-1, 1)

linear_reg_model.fit(x_train, y_train)
```

```python
layer_2_train_pred_mlr = linear_reg_model.predict(x_train)
layer_2_test_pred_mlr  = linear_reg_model.predict(x_test)
```

```python
# Evaluate metrics
model_name = linear_reg_model

print('Evaluating the model on the training dataset')
r2_tr_mlr = evaluate(model_name,x_train,y_train)
print('')
print('Evaluating the model on the testing dataset')
r2_ts_mlr = evaluate(model_name,x_test,y_test)
```
